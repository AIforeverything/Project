```python
# making the dataframe of csv file names in the directory
directory="./"
csv_files=glob.glob(os.path.join(directory,"*.csv"))
print(csv_files)
fileName_Df=pd.DataFrame()
L=[]
for file in csv_files:
    filename = os.path.splitext(os.path.basename(file))[0]
    L.append(filename+".csv")
print(L) 
fileName_Df["files"]=L   
fileName_Df
```

```python
#Extracting all symbols from all files and removing duplicates using set
stockNamesSet=set()
stockNamesSetwithNS=set()
stockSymbols_df=pd.DataFrame()
stockSymbols_df["Symbols"]=[]
max=0
for i in range(1,len(L)):
    df=pd.read_csv(fileName_Df["files"][i])
    max=len(df["SYMBOL \n"][1:])
    for j in range(1,max):
        stockNamesSetwithNS.add(df["SYMBOL \n"][j]+".NS")
        stockNamesSet.add(df["SYMBOL \n"][j])
print(stockNamesSet)
print(stockNamesSetwithNS)
print(len(stockNamesSet))
stockNamesSetList=list(stockNamesSet)
stockNamesSetList=sorted(stockNamesSetList)
print(stockNamesSetList)
stockSymbols_df["Symbols"]=stockNamesSetList
stockSymbols_df.head()
```

```python
#checking the stock details
def get_info(ticker):
    stock=yf.Ticker(ticker)
    print(stock.info)
```

```python
def save_csv(stock_name):
    stock=yf.Ticker(stock_name)
    df=stock.history(period="20y")["Close"]
    file=stock_name.replace(".NS","")+".csv"
    df.to_csv("data/"+file)
```

```python
for stock in stockNamesSetwithNS:
    save_csv(stock)
```

```python
#combining all stocks data into a single dataframe
allStocks_df=pd.DataFrame()
date_purpose=pd.read_csv("data/VSTIND.csv")
allStocks_df["Date"]=date_purpose["Date"]
try:
    for i in stockNamesSet:
        dfs=pd.read_csv("data/"+i+".csv")
        allStocks_df=allStocks_df.merge(dfs,left_on="Date",right_on="Date",how="left")
        allStocks_df=allStocks_df.rename(columns={"Close":i})
               
except Exception as e:
    print("Error: ",e)

allStocks_df.info()
```
